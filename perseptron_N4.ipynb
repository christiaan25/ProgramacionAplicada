```python
x = (4, 8, 6)  # Tupla fija

# Encontrar w tal que sum(w[i]*x[i] for i in range(3)) == 10
# Hay infinitas soluciones, aquí mostramos algunas formas de encontrarlas

# Solución 1: w con un único elemento no cero
def encontrar_w_simple(x, target=10):
    soluciones = []
    for i in range(len(x)):
        w = [0, 0, 0]
        w[i] = target / x[i]
        soluciones.append(tuple(w))
    return soluciones

# Solución 2: w con valores arbitrarios que cumplan la ecuación
def encontrar_w_completa(x, target=10):
    # 4*w1 + 8*w2 + 6*w3 = 10
    # Vamos a dar valores arbitrarios a w2 y w3 y calcular w1
    w2, w3 = 1, 1  # Valores arbitrarios
    w1 = (target - 8*w2 - 6*w3) / 4
    return (w1, w2, w3)

# Ejemplos de uso
print("Soluciones con un elemento no cero:")
for sol in encontrar_w_simple(x):
    print(f"w = {sol}")

print("\nSolución completa:")
sol_completa = encontrar_w_completa(x)
print(f"w = {sol_completa}")

# Verificación
def verificar(x, w):
    y = tuple(a*b for a,b in zip(x,w))
    return sum(y)

print("\nVerificación:")
print(f"Para w = {sol_completa}, sum(y) = {verificar(x, sol_completa)}")
```

```python
import numpy as np

# Datos de entrada (3 ejemplos, 3 características)
xs = np.array([
    [0.3, 0.6, 0.8],
    [0.5, 0.2, 0.9],
    [0.1, 0.7, 0.4]
])

# Salida real (3 ejemplos, 2 salidas)
sombra = np.array([
    [0.5, 0.8],
    [0.6, 0.4],
    [0.3, 0.9]
])

# Pesos iniciales (2 neuronas, 3 entradas)
w = np.array([
    [0.6, 0.3, 0.5],
    [0.7, 0.5, 0.6]
])

def calcular_y_calc(xs, w):
    return xs @ w.T  # Producto punto: xs * w^T

y_calc = calcular_y_calc(xs, w)
error = calcula_error(sombra, y_calc)
print("Error actual:", error)  # Ejemplo: 0.183
```

```python
w_mejor = np.array([
    [0.7, 0.4, 0.6],  # Ajustados para reducir el error
    [0.8, 0.6, 0.7]
])

y_calc_mejor = calcular_y_calc(xs, w_mejor)
error_mejor = calcula_error(sombra, y_calc_mejor)
print("Error mejorado:", error_mejor)  # Debería ser < 0.183
```

Retropropagacion

```python
import copy
import numpy as np

# Datos de ejemplo CORRECTAMENTE dimensionados
cubo = [
    [0.3, 0.6, 0.8],  # 3 entradas
    [0.5, 0.2, 0.9],
    [0.1, 0.7, 0.4]
]

sombra = [
    [0.5, 0.8],  # 2 salidas
    [0.6, 0.4],
    [0.3, 0.9]
]

# Inicialización CORRECTA de w (2 salidas x 3 entradas)
w = [
    [0.6, 0.3, 0.5],  # Pesos para salida 1
    [0.7, 0.5, 0.6]   # Pesos para salida 2
]

# --- Funciones corregidas ---
def calcula_salida(x, w):
    y = []
    for i in range(len(w)):  # Para cada neurona de salida
        yi = sum(w[i][j] * x[j] for j in range(len(x)))
        y.append(yi)
    return y

def calcula_salidas(xs, w):
    return [calcula_salida(x, w) for x in xs]

def calcula_error(y_a, y_b):
    error = 0
    for j in range(len(y_a)):
        for i in range(len(y_a[0])):
            error += (y_a[j][i] - y_b[j][i])**2
    return error / (len(y_a) * len(y_a[0]))

def backprop_lin(x, w_ant, y_esp, y_calc, va=0.1):
    w_sig = copy.deepcopy(w_ant)
    for i in range(len(w_sig)):
        for j in range(len(w_sig[0])):
            w_sig[i][j] += va * x[j] * (y_esp[i] - y_calc[i])
    return w_sig

# --- Ejecución segura ---
x = cubo[0]
ye = sombra[0]
yc = calcula_salida(x, w)

w_calc = backprop_lin(x, w, ye, yc, va=1)
error = calcula_error(calcula_salidas(cubo, w_calc), sombra)

print("Pesos ajustados:", w_calc)
print("Error:", error)
```

Generalizacion de retropropagacion

```python
import copy

def entrenamiento(xs,w,y_esp,va=0.1, epocas=1, dibujar=True):
  w1=w
  for k in range(epocas):
    y_calc=calcula_salidas(cubo,w1)
    print('Epoca:', k, 'error:', calcula_error(y_calc,sombra))

    if dibujar==True:
      dibuja(y_calc)

    for i in range(len(xs)):
      x=xs[i]
      ye=y_esp[i]
      yc=calcula_salida(cubo[i], w1)
      w1=backprop_lin(x,w1,ye,yc,va)
  return w1

w=[[w11 ,w12 , w13], [w21, w22, w23]]

w_final=entrenamiento(cubo,w,sombra,0.5, 15)

print('w_final=', w_final)
```
